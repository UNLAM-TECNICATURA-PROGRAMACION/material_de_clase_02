<u>Operadores lógicos</u>
<br>AND y OR se utilizan para conectar dos condiciones o valores de verdad. En el caso del NOT se aplica sobre un valor de verdad.
<br><br>AND: Es necesario que se cumplan ambas condiciones para que el resultado sea verdadero.
<br>OR: Con que se cumpla alguna de las condiciones, alcanza para que el resultado sea verdadero.
<br>NOT: Invierte el valor de verdad de la condición sobre la cual se aplica. Con lo que si originalmente la condición era Verdadera, despues del NOT será falsa y viceversa.

```python
#AND (Conector y) - Retorna True cuando ambos valores son True
#Esto sería cua
print( True and True )
print( False and True )
print( True and False )
print( False and False )

#Por ejemplo, si quisieramos saber si el jugador ingreso una opción válida
#comprendida entre 1 y 4, la condición sería la siguiente
opcion = 1
print("Si esto retorna True, la opción es válida: ", opcion >= 1 and opcion <= 4 )
```

```python
#OR (Conector ó) - Retorna True cuando alguno de los valores es True
print( True or True )
print( False or True )
print( True or False )
print( False or False )

#Por ejemplo, supongamos un juego donde el game over puede darse por alguna de
#las siguientes dos condiciones:
#A) Que el jugador pierda todas sus vidas
#ó
#B) Que el tiempo para completar el nivel haya expirado
#La condición sería algo así:
vidas = 3
tiempo = 60
print("Si esto retorna True, es Game Over:", vidas == 0 or tiempo == 0 )
```

```python
#NOT (Negador) - Invierte el valor de verdad dado
print( not True )
print( not False )

#Por ejemplo, si en un juego utilizamos una variable para controlar si el jugador
#está vivo o muerto.
#El negador se podría aplicar de la siguiente manera:
muerto = False #Inicialmente está vivo, con lo cual es Falso que esté muerto
print("Si esto retorna True, está vivo:", not muerto )

#Esto se puede utilizar en un bucle por ejemplo, realizando acciones mientras
#el jugador no esté muerto.
```

<br><br>En programación hay una serie de estructuras de control que son prácticamente la base de la programación. Estamos hablando de los condicionales y los bucles en sus distintas formas.
<br><br><u>Condicionales</u>
<br>Sirven para ejecutar bloques de código dependiendo de si una condición es verdadera o falsa.
<br><br>Diagrama de flujo

 grabacion_teams.png

```python
#Por ejemplo, supongamos un RPG con combates autómaticos donde en cada turno
#nuestro personaje usará magia si dispone de al menos 10 puntos de MP y sino
#atacará con melee.

MP_NEEDED = 10
player_mp = 100

#La parte del código que controlará dicha acción será un condicional, algo así:
if player_mp >= MP_NEEDED:
    print("Atacar con magia")
else:
    print("Atacar con melee")

#La sentencia if necesita de una o varias condiciones para saber que camino elegir
#en este caso la condición es:
#player_mp >= MP_NEEDED

#Si la condición se cumple, el programa ejecuta las sentencias de código que se
#encuentran en la parte verdadera (el bloque debajo de la sentencia if)

#Si la condición no se cumple, el programa sigue con el bloque de código debajo
#de la sentencia else

#La parte falsa (else) es opcional, es decir que si no necesitamos hacer nada
#en caso de que no se cumpla la condición, no es necesario incluirla
```

<br><u>Condicional de selección múltiple</u>
<br><br>Hay otro condicional muy utilizado, que es el de selección múltiple. A partir de Python 3.10, disponemos de match, que permite comparar un valor contra varios "patrones" y elegir el camino que corresponda en caso de que exista coincidencia.
<br><br>Diagrama de flujo

image.png

```python
#Por ejemplo; supongamos que el jugador debe elegir un personaje y dependiendo
#de su elección se cargarán distintos valores en sus atributos iniciales

#El tipo especial None, en python representa la ausencia de un valor (nada)
character = None

#Utilizaremos la función input() para que el jugador pueda ingresar una opción
#Tener en cuenta que la función input retorna una cadena aunque hayamos ingresado
#un número, con lo cual hay que hacer lo que se conoce como cast (conversión)
#para que option almacene un entero y el match pueda funcionar bien.
option = int(input("Elija su aventurero (1 a 3):"))

#La parte del código que controlará dicha acción podría ser un condicional de
#selección múltiple:
match option:
    case 1:
        character = "Legolas"
        player_atq = 10
        player_def = 5
    case 2:
        character = "Aragorn"
        player_atq = 15
        player_def = 10
    case 3:
        character = "Gimli"
        player_atq = 5
        player_def = 5
    case _: #Cualquier otra opción que no sea 1, 2 ó 3
        print("Opción no válida")

#Si character es distinto de None es porque se eligió un personaje
#entonces lo muestro por pantalla
if character != None:
    print( f"NOMBRE: {character}   ATQ: {player_atq:02}   DEF: {player_def:02}" )
```

```python
#En el ejemplo anterior, si ingresamos una cadena en lugar de un número, el input
#arrojará un error. Para manejar este tipo de errores durante la ejecución se suele
#utilizar lo que se conoce como try/except en Python (try/catch en otros lenguajes)

#La sintáxis básica sería así:
#try:
#     Código que puede causar un error
#except TipoDeError:
#     Código que se ejecuta si ocurre ese error

#Básicamente es decir, ejecutá este código y si genera error hacé esto otro.

character = None
try:
    option = int(input("Elija su aventurero (1 a 3):"))
except ValueError:
    option = None

match option:
    case 1:
        character = "Legolas"
        player_atq = 10
        player_def = 5
    case 2:
        character = "Aragorn"
        player_atq = 15
        player_def = 10
    case 3:
        character = "Gimli"
        player_atq = 5
        player_def = 5
    case _:
        print("Opción no válida")

if character != None:
    print( f"NOMBRE: {character}   ATQ: {player_atq:02}   DEF: {player_def:02}" )
```

<br><u>Bucles</u>
<br><br>Los bucles permiten repetir la ejecución de un bloque de código mientras se cumpla una condición.
<br><br>
Ciclo For
<br>Diagrama de flujo
<br>
image.png

```python
#Un ciclo for se suele utilizar cuando sabemos cuantas veces vamos a iterar.
#Por ejemplo, supongamos que tenemos que elegir con que arma equipar a cada
#personaje de nuestro grupo de 5 en total.

#Podriamos utilizar un ciclo for, ya que la acción que se va a realizar será
#la misma para cada personaje y sabemos que tendremos que realizarla 5 veces.

for i in range(5): #range(5) genera la secuencia 0,1,2,3,4
    arma = input("Ingrese nombre de arma ")
    print(f"Personaje {i+1} se equipó con {arma}")
```

```python
#La sentencia for también se puede utilizar para recorrer un conjunto de
#elementos, aunque no sepamos cuantos son. Por ahora haremos un ejemplo sencillo.

#El siguiente ejemplo recorre el conjunto de caracteres de la cadena y los imprime
for letra in "Palabra":
    print(letra)
```

Ciclo While
<br>Diagrama de flujo
<br>
image.png

```python
from random import randint
from time import sleep
#Un ciclo while se suele utilizar cuando queremos repetir un bloque de código
#mientras se cumpla una condición, lo que no sabemos bien es cuantas veces
#vamos a iterar.

#Por ejemplo, supongamos dos personajes que luchan entre sí y mientras ninguno
#muera seguiran luchando. Una forma de plantear el código para representar
#dicho escenario podría ser con un ciclo while.

#Agregamos un delay de medio segundo para que cada pasada del ciclo se frene un poco
#y nos permita ver los mensajes de manera mas pausada
DELAY = 0.5
HP_QUITADO = 10
ryu_hp = 50
ken_hp = 50

#La condición se cumplirá hasta que alguno o ambos mueran
while ryu_hp != 0 and ken_hp != 0:
    nro = randint(1,3)
    match nro:
        case 1:
            print("Ataca RYU")
            ken_hp-=HP_QUITADO
        case 2:
            print("Ataca KEN")
            ryu_hp-=HP_QUITADO
        case 3:
            print("Ambos se golpean al mismo tiempo")
            ken_hp-=HP_QUITADO
            ryu_hp-=HP_QUITADO
    sleep(DELAY)

if ryu_hp != 0:
    print("Ganó RYU")
else:
    if ken_hp != 0:
        print("Ganó KEN")
    else:
        print("Double KO")

#Con randint generamos un numero aleatorio entre 1 y 3 para simular el accionar de los
#personajes de manera aleatoria.
```
